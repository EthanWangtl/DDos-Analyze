```python
import numpy as np
import pandas as pd
```

```python
data1 = pd.read_csv('LDAP.csv')
data2 = pd.read_csv('MSSQL.csv')
data3 = pd.read_csv('NetBIOS.csv')
data4 = pd.read_csv('Portmap.csv')
data5 = pd.read_csv('Syn.csv')
data6 = pd.read_csv('UDP.csv')
data7 = pd.read_csv('UDPLag.csv')
```

```python
data1_sample = data1.sample(500000)
data2_sample = data2.sample(500000)
data3_sample = data3.sample(500000)
data5_sample = data5.sample(500000)
data6_sample = data6.sample(500000)
data7_sample = data7.sample(500000)
```

```python
data = pd.concat([data1_sample,data2_sample,data3_sample,data4,data5_sample,data6_sample,data7_sample], ignore_index = True)
```

```python
data[' Label'].value_counts()
```

```python
data = data.drop(['Unnamed: 0'],axis = 1)
data = data.drop(['SimillarHTTP'],axis = 1)
```

```python
data_real = data.replace(np.inf, np.nan)
data_real.isnull().sum().sum()
```

```python
data_df = data_real.dropna(axis=0)
data_df.isnull().sum().sum()
```

```python
df = data_df.copy()
cat_columns = df.select_dtypes(['object']).columns
cat_columns
```

```python
df[cat_columns] = df[cat_columns].astype('category')
df[cat_columns] = df[cat_columns].apply(lambda x: x.cat.codes)
df
data = df
data.dtypes
```

```python
data[' Label'].value_counts()
```

```python
data= data.sample(n=len(data), random_state=1)
```

```python
X = data.iloc[:, 0:-1].values
y = data.iloc[:, -1].values
print('length of the dataset now', len(X))
```

```python
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
X= standardScaler.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.67, test_size = 0.33, random_state = 42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#NN
```

```python
#CNN algorithm libraries
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Conv1D, GlobalMaxPooling1D
from tensorflow.keras.optimizers import RMSprop

#for evaluation of trained algorithms 
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
model = Sequential()
```

```python
model.add(Dense(64, input_dim=85))
model.add(Activation('relu'))
model.add(Dropout(0.01))
```

```python
model.add(Dense(32))
model.add(Activation('relu'))
model.add(Dropout(0.01))
model.add(Dense(8, activation="softmax"))
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['acc'])
```

```python
model.summary()
model.compile(loss='sparse_categorical_crossentropy',optimizer=RMSprop(),metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,batch_size=20,epochs=20,
          validation_split=0.1)
```

```python
y_pred = model.predict(X_test)
y_pred=y_pred.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
rf_conf_mat = confusion_matrix(y_test, y_pred)
print("CNN Confusion: \n", rf_conf_mat)
```

```python
print('\n')
print("Precision, Recall, F1")
print('\n')
CR=classification_report(y_test, y_pred)
print(CR)
print('\n')
```
